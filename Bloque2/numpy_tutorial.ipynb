```python
#Definimos la función utilizando math:
import math

def sigmoide_math(x):
    '''
    Input: elemento del tipo int o float
    output: elemento del tipo int o float
    '''
    return 1/(1 + math.exp(-x))
```

En Deep Learning las entradas de la función sigmoide suelen ser vectores, probemos con enteros, flotantes y vectores:

```python
#Calculamos el valor de sigmoide para x = 1 como entero y flotante:

x = int(1)
y = float(1)

print("x es del tipo {} y sigmoide(x) tiene el valor {:.5f} y es del tipo {} y sigmoide(y) tiene el valor {:.5f}.".format(type(x), sigmoide_math(x),type(y), sigmoide_math(y)) )
#print("database es una {} con {} elements and shape {}.".format(type(database), len(database), database.shape))
```

<font size = 4>

Ahora bien, si intentamos calcular el valor de **sigmoide_math()** de [X = np.array([1,2,3])](https://numpy.org/doc/stable/reference/generated/numpy.array.html?highlight=array#numpy.array), obtenemos:

```python
import numpy as np

X = np.array([1,2,3]) #X como vector
sigmoide_math(X)  #Veremos que nos devuelve un error
```

<font size = 4>

Este error se debe a que **math** funciona con variables del tipo entero y flotante.

Probemos utilizando **np.exp()** para los tres tipo de variable: entero, flotante, vector(arreglo):

```python
#Definimos la función sigmoid utilizando Numpy:

def sigmoid_np(x):
    """
    Input: Vector X del tipo arreglo de numpy
    Output: Arreglo de numpy
    """
    return 1/(1 + np.exp(-x))
```

```python
#Evaluamos con x, y, X:

print("'x' es del tipo {} y sigmoide(x) tiene el valor {:.5f}. \n 'y' es del tipo {} y sigmoide(y) tiene el valor {:.5f}. \n 'X' es del tipo {} y sigmoid de X tiene el valor de {}." \
      .format(type(x), sigmoid_np(x),type(y), sigmoid_np(y), type(X), sigmoid_np(X)))
```

<font size = 4>

Lo que ocurre es que **Numpy** calcula la exponencial de cada una de las entradas del vector 'X' y regresa np.exp(X) como un vector de la misma forma que 'X'.

Operaciones con arrays

Los arrays permiten hacer operaciones aritméticas básicas entre ellos en la forma que uno esperaría que
se hicieran, es decir, haciéndolo elemento a elemento; para ello ambos arrays deben tener siempre la
misma longitud, por ejemplo:


```python
x = array([5.6, 7.3, 7.7, 2.3, 4.2, 9.2])
print(x+decimales)
```

```python
print(x*decimales)
```

```python
print(x/decimales)
```

Como podemos ver las operaciones se hacen elemento a elemento, por lo que ambas deben tener la misma forma (shape()). Fíjense que en la división el resultado del primer elemento es indefinido/infinito
(Inf) debido a la división por cero.
Varios arrays se pueden unir con el método [np.concatenate()](https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html?highlight=concatenate#numpy.concatenate), que también se puede usar para
añadir elementos nuevos:


```python
z = np.concatenate((x, decimales))
print(z)
```

```python
z = np.concatenate((x, [7]))
print(z)
```

Es importante fijarse que los arrays o listas a unir deben darse como un iterable (tupla, lista, array) como
por ejemplo (x, [7]), (x, [2,4,7]) o (x, array([2,4,7])).
Para añadir elementos, numpy tiene las funciones insert() y append(), que funcionan de manera
similar a sus equivalentes en listas, pero en este caso son funciones y no métodos que se aplican a un
array, si no que el array en cuestión hay que darlo como parámetro:

```python
# Añadimos el elemento 100 al array z, al final
z = np.append(z, 100)
print(z)
```

```python
# Añadimos el elemento 200 al array z, en el tercer puesto (índice 2)
z = np.insert(z, 2, 200)
print(z)
```

Como se ve, a diferencia de las listas, el primer parámetro es el array y luego el elemento que se quiere añadir, en el caso de [append()](https://numpy.org/doc/stable/reference/generated/numpy.append.html?highlight=append#numpy.append) y el array, la posición y luego elemento a añadir en el caso de
append(). Esto se debe a que estas funciones devuelven una copia del array sin modificar el original como hacen los métodos de listas correspondientes. Si en lugar de un elemento a insertar se da una
lista y otro array, añade todos los elementos de la lista (a append() habría que dar también una lista
de posiciones, como segundo parámetro).


```python
# Creamos un nuevo array anhadiendo al array z,
# los elementos -10, -20, -30 en las posiciones con indices 2, 4, -1,respectivamente
```

```python
y = np.insert(z, [2, 4, -1], [-10, -20, -30])
print(y)
```

Arrays multidimensionales

Hasta ahora sólo hemos trabajado con arrays con una sola dimensión, pero numpy permite trabajar con
arrays de más dimensiones. Un array de dos dimensiones podría ser por ejemplo un array que tuviera
como elementos un sistema de ecuaciones o una imagen. Para crearlos podemos hacerlo declarándolos
directamente o mediante funciones como [np.zeros()](https://numpy.org/doc/stable/reference/generated/numpy.zeros.html?highlight=zeros#numpy.zeros) o [np.ones()](https://numpy.org/doc/stable/reference/generated/numpy.zeros.html?highlight=zeros#numpy.zeros) dando como parámetro una
tupla con la forma del array final que queramos; o también usando [np.arange()](https://numpy.org/doc/stable/reference/generated/numpy.arange.html?highlight=arange#numpy.arange) y crear un array
unidimensional y luego cambiar su forma. Veamos algunos ejemplos:

```python
# Array de 3 filas y tres columnas, creado implícitamente
arr0 = np.array([[10,20,30],[9, 99, 999],[0, 2, 3]])
print(arr0)
```

```python

# Array de ceros con 2 filas y 3 columnas
arr1 = np.zeros((2,3))
print(arr1)
```

```python
# Array de unos con 4 filas y una columna
arr2 = ones((4,1))
print(arr2)
```

```python
# Array unidimensional de 9 elementos y cambio su forma a 3x3
arr3 = np.arange(9).reshape((3, 3))
print(arr3)
```

```python
arr2.shape
```

Como vemos en la última línea, la forma o shape() de los arrays se sigue dando como una tupla, con

la segunda dimensión o eje es una columna. Es por eso que al usar las funciones zeros(), ones(),
reshape(), etc. hay que asegurarse que el parámetro de entrada es una tupla con la longitud de cada
eje. Cuando usamos la función len() en un array bidimensional, el resultado es la longitud del primer
eje o dimensión, es decir, len(arr2) es 4.
El acceso a los elementos es el habitual, pero ahora hay que tener en cuenta el eje al que nos referimos;
además podemos utilizar ”:” como comodín para referirnos a todo el eje. Por ejemplo:

```python
# Primer elemento de la primera fila y primera columna (0,0)
arr0[0,0]
```

```python
# Primera columna
arr0[:,0]
```

```python
# Primera fila
arr0[0,:]
```

```python
# Elementos 0 y 1 de la primera fila
arr0[0,:2]
```

Filtros y máscaras de arrays. Arrays enmascarados

Una de las mejores utilidades de numpy es trabajar con índices y máscaras de datos para limitar o
seleccionar parte de los datos. Supongamos que tenemos un array de datos, pero que solo nos interesa
los positivos, que queremos manipular después. Hay varias formas de seleccionarlos definiendo un array
máscara con la condición que nos interesa:


```python
datos = array([3, 7, -2, 6, 7, -8, 11, -1, -2, 8])
datos
```

```python
mask = datos >= 0
mask
```

```python
datos*mask
datos[mask]
```

Usando un array mask de booleanos, podemos operar con el el array de datos, cuando un valor se
multiplica por True es equivalente a multiplicarse por 1 y si es con False, a multiplicarse por 0.
Por eso el resultado es un array del mismo tamaño, pero los elementos que no cumplen la condición se
hacen 0.

Si por el contrario usarmos usamos mask como un array de índices, el resultado es un array con los
elementos cuyo índice corresponda con True, ignorando los de índice False. Usaremos uno u otro
según lo que queramos hacer, el truco consiste es crear de manera correcta la máscara de datos.

Veamos el caso de un array 2D con dos columnas, pero queremos limitar todos los datos en criterios
en las dos columnas. Primero creamos una máscara como producto de las dos condiciones, y luego la
usamos como array de índices en el array original:


```python
from numpy import random
datos2 = random.randint(-10, 20, (10,2))
datos2
```

```python
# Solo queremos los datos que el la columna 0 sean mayores
# que 0 pero menores que 10 en la columna 1
condicion1 = datos2[:,0] > 0
condicion1 = datos2[:,1] < 10
mask_col0 = condicion1*condicion1
mask_col0

```

```python
datos2[mask_col0]
```

Como se ve, el resultado en un array de dos columna, donde en la primera columna son todos positivos
y en la segunda menores que +10. ¿Y si queremos que al menos se cumpla una condición? Simplemente
tenermos que sumar las dos máscaras (las de cada columna) en lugar de multiplicarla, básicamente es
como multiplicar o sumar unos o ceros (True o False).


Como este tipo de operaciones tienen mucho potencial y pueden llegar a se complejas. numpy tiene un
módulo que puede ayudar en estos casos, que es el de arrays enmascarados ([numpy.ma](https://numpy.org/doc/stable/reference/maskedarray.generic.html?highlight=ma#module-numpy.ma)). Se trata de un
tipo de datos que permite ignorar algunos elementos de un array según ciertas condiciones. Vemos un
ejemplo:


```python
import numpy.ma as ma
x = array([1, 2, 3, -1, 5])
```

```python
# Enmascaramos en cuarto elemento
mx = ma.masked_array(x, mask=[0, 0, 0, 1, 0])
print(mx.mean())
```

```python
print(x.mean())
```

Como se ve, el array enmascarado ha ignorado el valor que se enmascara con True dando un resultado
distinto en la media, pero no lo ha eliminado del array

```python
x.view(ma.MaskedArray)
```

```python
mx.view(ma.MaskedArray)
```

Podemos usar algunas funciones de ma para crear las máscaras, como masked_greater(),
masked_inside(), masked_where(), etc.

```python
a = np.arange(4)
print(a)
```

```python
ma.masked_where(a <= 2, a)
```

Arrays estructurados


Aunque los arrays pueden contener cualquier tipo de dato, los arrays normales sólo pueden ser de un
único tipo. Para esto extiste una variante de arrays para contenidos complejos o estructurados, llamado structured arrays, que permiten tratar arrays por estructuras o por campos de estruturas. Además de
poder contener distintos tipos de datos, facilitan el acceso por columnas. Vemos un ejemplo con un
array con distintos tipos de datos:

```python
 # Array estucturado de 5 elementos, vacio
galaxies = np.zeros(5, dtype = {'names': ('name', 'order', 'type', 'magnitude'),'formats': ('U16', 'i4', 'U10', 'f8')})
# Listas de datos para contruir el array estructurado
names = ["M 81", "NGC 253", "M 51", "NGC 4676", "M 106"]
types = ["SA(s)b", "SAB(s)c", "Sc", "Irr", "SAB(s)bc"]
magnitudes = [6.93, 7.1, 8.4, 14.7, 9.1]
order = list(range(5))
```

```python
# Anhadimos valores a los campos (columnas)
galaxies['name'] = names
galaxies['type'] = types
galaxies['magnitude'] = magnitudes
galaxies['order'] = order
print(galaxies)
```

Se trata de un array con cinco entradas (o records) y cada una de ellas posee cuatro campos de distinto
tipo, indicados con la propiedad dtype. En este caso son un string unicode de longitud máxima 16
(U16), un entero 4 bytes (i.e. 32 bit) (i4), string unicode de longitud máxima 16 (U16) y un float de
4 bytes (i.e. 64 bit). El dtype de numpy describe cómo interpretar cada elemento en bytes de bloques
de memoria fijos. No sólo se trata de si son float, int, etc., el dtype describe lo siguiente:

Tipo de dato (int, float, objeto Python, etc.)
Tamaño del dato (cuantos bytes puede ocupar)
Orden de bytes de datos (little-endian o big-endian)
Si son datos estructurado (por ejemplo mezcla de tipos de dato), también:
• Nombre de los campos
• Tipo de dato de cada campo
• Qué parte del bloque de memoria ocupa cada campo
• Si en dato es un sub-array, su forma y tipo de dato
De manera resumida, para definir el tipo de cada elemento podemos usar una de las siguientes cadenas:

b1, i1, i2, i4, i8, u1, u2, u4, u8, f2, f4, f8, c8, c16, a<n>


que representan, respectivamente, bytes, ints, unsigned ints, floats, complex y strings de longitud fija.
También se pueden usar los tipos de datos estándar de Python equivalentes (int, float, etc.)

Teniendo un array estructurado como el anterior, podemos ver cada elemento haciendo el indexado
habitual y también por campos (columnas):

```python
print(galaxies)
```

```python
# Columna 'name' del array
galaxies['name']
```

```python
# Primer elemento del array, con todos los campos (columnas)
galaxies[0]
```

En este sentido, calculemos el **gradiente** de la función **sigmoide** el cual se determina mediante la ecuación: $$ sigmoid\_derivative (x) = \sigma '(x) = \sigma (x) (1 - \sigma (x)) \tag {2} $$
    
Puedes comprobar que la ecuación es correcta diferenciando respecto a la variable de la función.

```python
#Definimos la función:

def sigmoid_gradient(x):
    """
    Input: x, un entero, flotante o arreglo de numpy.
    output: gradiente de sigmoid(x), flotante.
    """
    sig = sigmoid_np(x)
    sig_grad = sig*(1 - sig)
    return sig_grad
```

Implementemos la función **np.reshape** reordenando un arreglo de la forma (3,3,2) en un arreglo de la forma (18,1):

```python
#Definimos la función Reshape:

def image2vector(image):
    """
    Input:
    image -- arreglo de numpy de la forma (length, height, depth)
    
    Returns:
    v -- vector de la forma (length*height*depth, 1)
    """
    
    v = image.reshape((image.shape[0]*image.shape[1]*image.shape[2]),1)
    
    return v
```

6.- Broadcast y normalización

<font size = 4>

Un proceso fundamental en algoritmos de Deeplearning esla normalización de los datos con lo que se entrenará una red neuronal.
Por ejemplo, es posible normalizar una matriz o arreglo a partir de normalizar cada uno de sus renglones o filas.
Esto es, dada la matriz $$x:
\begin{bmatrix}
    0 & 3 & 4 \\
    2 & 6 & 4 \\
\end{bmatrix}\tag{1}$$ entonces $$\| x\| = \begin{bmatrix}
    5 \\
    \sqrt{56} \\
\end{bmatrix}\tag{2} $$ y        $$ x\_normalized = \frac{x}{\| x\|} = \begin{bmatrix}
    0 & \frac{3}{5} & \frac{4}{5} \\
    \frac{2}{\sqrt{56}} & \frac{6}{\sqrt{56}} & \frac{4}{\sqrt{56}} \\
\end{bmatrix}\tag{3}$$

<font size = 4>

Implementemos una función que nos permita llevar acabo la operación, haremos uso del atributo de numpy **np.linalg.norm(x)**.
    

```python
#Función que normaliza cada renglón de un arreglo:

def normalizeRows(x):
    """
    Input:
    x -- Un arreglo de numpy de la forma (n, m)
    
    Output:
    x -- El arreglo x con cada uno de sus renglones normalizado.
    """
    x_norm = np.linalg.norm(x, axis = 1, keepdims = True)
    
    # Broadcast
    x = x/x_norm

    return x
```

```python
#Probamos la función:

x = np.array([
    [0, 3, 4],
    [1, 6, 4]])

print("normalizeRows(x) = " + str(normalizeRows(x)))
```

Al proceso de efectuar operaciones entre arreglos de diferentes dimesiones o escalares y arreglos se le conoce como **Broadcasting**. Y es uno de los procesos cotidianos en Machine Learning y Deep Learning.

<font size=4> Puedes entender mejor el Broadcasting [aquí](https://numpy.org/doc/stable/user/basics.broadcasting.html)

<font size = 4>

Como un ejercicio extra de broadcasting implementemos la función **Softmax**, esta se define como:
    
    
- $ \text{Sea } x \in \mathbb{R}^{1\times n} \text{,       } softmax(x) = softmax(\begin{bmatrix}
    x_1  &&
    x_2 &&
    ...  &&
    x_n  
\end{bmatrix}) = \begin{bmatrix}
     \frac{e^{x_1}}{\sum_{j}e^{x_j}}  &&
    \frac{e^{x_2}}{\sum_{j}e^{x_j}}  &&
    ...  &&
    \frac{e^{x_n}}{\sum_{j}e^{x_j}} 
\end{bmatrix} $ 

- $\text{Para un arreglo } x \in \mathbb{R}^{m \times n} \text{,  $x_{ij}$ asigna al elemento en el $i^{th}$ renglon y $j^{th}$ columna de $x$, entonces tenemos: }$  $$softmax(x) = softmax\begin{bmatrix}
    x_{11} & x_{12} & x_{13} & \dots  & x_{1n} \\
    x_{21} & x_{22} & x_{23} & \dots  & x_{2n} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    x_{m1} & x_{m2} & x_{m3} & \dots  & x_{mn}
\end{bmatrix} = \begin{bmatrix}
    \frac{e^{x_{11}}}{\sum_{j}e^{x_{1j}}} & \frac{e^{x_{12}}}{\sum_{j}e^{x_{1j}}} & \frac{e^{x_{13}}}{\sum_{j}e^{x_{1j}}} & \dots  & \frac{e^{x_{1n}}}{\sum_{j}e^{x_{1j}}} \\
    \frac{e^{x_{21}}}{\sum_{j}e^{x_{2j}}} & \frac{e^{x_{22}}}{\sum_{j}e^{x_{2j}}} & \frac{e^{x_{23}}}{\sum_{j}e^{x_{2j}}} & \dots  & \frac{e^{x_{2n}}}{\sum_{j}e^{x_{2j}}} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    \frac{e^{x_{m1}}}{\sum_{j}e^{x_{mj}}} & \frac{e^{x_{m2}}}{\sum_{j}e^{x_{mj}}} & \frac{e^{x_{m3}}}{\sum_{j}e^{x_{mj}}} & \dots  & \frac{e^{x_{mn}}}{\sum_{j}e^{x_{mj}}}
\end{bmatrix} = \begin{pmatrix}
    softmax\text{(first row of x)}  \\
    softmax\text{(second row of x)} \\
    ...  \\
    softmax\text{(last row of x)} \\
\end{pmatrix} $$

```python
#Función Softmax:

def softmax(x):
    """
    Input:
    x -- Un arreglo con dimensiones (n,m)

    Output:
    s -- Un arreglo igual a la función softmax de x con dimensiones (n,m)
    """
    
    x_exp = np.exp(x)
    
    x_sum = np.sum(x_exp, axis = 1, keepdims = True)
    
    s = x_exp/x_sum #Broadcast
    
    return s
```

```python
#Calculamos la función softmax de un arreglo:

x = np.array([[9, 2, 5, 0, 0], [7, 5, 0, 0 ,0]])

print("x es un arreglo de dimensiones {} y softmax(x) es un arreglo con dimensiones {}. \nPor su parte softmax(x) tiene los valores: \n{}"\
     .format(x.shape, softmax(x).shape, softmax(x)))
```

7.- Vectorización

<font size=4>
    
Como se ha mencionado, en deep learning se manejan conjuntos de datos muy grandes. De manera que una función no computacionalmente óptima puede convertirse en un gran cuello de botella en el algoritmo y puede dar como resultado un modelo que demore años en ejecutarse. Para asegurarse de que el código sea computacionalmente eficiente se utiliza la vectorización. 
Por ejemplo, comparemos entre las siguientes implementaciones del producto **dot/ external/ elementwise**.

```python
import time

x1 = [9, 2, 5, 0, 0, 7, 5, 0, 0, 0, 9, 2, 5, 0, 0]
x2 = [9, 2, 2, 9, 0, 9, 2, 5, 0, 0, 9, 2, 5, 0, 0]

#Implementación del producto punto clásico entre vectores

tic = time.process_time()
dot = 0
for i in range(len(x1)):
    dot+= x1[i]*x2[i]
toc = time.process_time()
print ("Producto punto = " + str(dot) + "\n ----- Tiempo de cá = " + str(1000*(toc - tic)) + "ms")

# Implementación del producto externo

tic = time.process_time()
outer = np.zeros((len(x1),len(x2))) # Definimos un arreglo len(x1)*len(x2) con únicamente ceros.
for i in range(len(x1)):
    for j in range(len(x2)):
        outer[i,j] = x1[i]*x2[j]
toc = time.process_time()
print ("Producto externo = " + str(outer) + "\n ----- Tiempo de cálculo = " + str(1000*(toc - tic)) + "ms")

# Implementación del producto Elementwise

tic = time.process_time()
mul = np.zeros(len(x1))
for i in range(len(x1)):
    mul[i] = x1[i]*x2[i]
toc = time.process_time()
print ("Producto Elementwise = " + str(mul) + "\n ----- Tiempo de cálculo = " + str(1000*(toc - tic)) + "ms")

# Implementación del producto punto general

W = np.random.rand(3,len(x1)) # Arreglo de dimensión 3*len(x1)
tic = time.process_time()
gdot = np.zeros(W.shape[0])
for i in range(W.shape[0]):
    for j in range(len(x1)):
        gdot[i] += W[i,j]*x1[j]
toc = time.process_time()
print ("gdot = " + str(gdot) + "\n ----- Tiempo de cálculo = " + str(1000*(toc - tic)) + "ms")
```

<font size=4>
    
Comparemos con las mismas operaciones utilizando la vectorización de Numpy:

```python
x1 = [9, 2, 5, 0, 0, 7, 5, 0, 0, 0, 9, 2, 5, 0, 0]
x2 = [9, 2, 2, 9, 0, 9, 2, 5, 0, 0, 9, 2, 5, 0, 0]

#Producto punto entre vectores de numpy

tic = time.process_time()
dot = np.dot(x1,x2)
toc = time.process_time()
print ("Producto punto = " + str(dot) + "\n ----- Tiempo de cálculo = " + str(1000*(toc - tic)) + "ms")

#Producto externo entre vectores de numpy:

tic = time.process_time()
outer = np.outer(x1,x2)
toc = time.process_time()
print ("Producto externo = " + str(outer) + "\n ----- Tiempo de cálculo = " + str(1000*(toc - tic)) + "ms")

#Producto elementwise entre vectores

tic = time.process_time()
mul = np.multiply(x1,x2)
toc = time.process_time()
print ("elementwise = " + str(mul) + "\n ----- Tiempo de cálculo = " + str(1000*(toc - tic)) + "ms")

#Producto punto general entre vectores

tic = time.process_time()
dot = np.dot(W,x1)
toc = time.process_time()
print ("gdot = " + str(dot) + "\n ----- Tiempo de cálculo = " + str(1000*(toc - tic)) + "ms")
```

<font size=4>


Como se puede observar, la implementación vectorizada es mucho más limpia y eficiente. Para vectores (arreglos) más grandes, las diferencias en el tiempo de cálculo se vuelven aún mayores.

Obsérvese que [`np.dot()`](https://numpy.org/doc/stable/reference/generated/numpy.dot.html?highlight=dot#numpy.dot) realiza una multiplicación matriz-matriz o matriz-vector. Esto es diferente de [np.multiply()](https://numpy.org/doc/stable/reference/generated/numpy.multiply.html?highlight=multiply) y el operador `*` (que es equivalente a `. *` En Matlab / Octave), que realiza una multiplicación por elementos.
